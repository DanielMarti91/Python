```python
import h2o
from h2o.estimators import H2ORandomForestEstimator
```

```python
h2o.init()
```

```python
# Cargar data CSV
data = h2o.import_file('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
```

```python
data.col_names
```

```python
# predictores del modelo
training_columns = ['C1', 'C2', 'C3', 'C4']
# target del modelo
response_column = 'C5'
```

```python
# Construcción de datos de entrenamiento y prueba
train, test = data.split_frame(ratios=[0.8])
```

```python
# definición del modelo
model = H2ORandomForestEstimator(ntrees=50, max_depth=20, nfolds=10)
```

```python
# modelo de entranamiento
model.train(x=training_columns, y=response_column, training_frame=train)
```

```python
model
```

```python
# Performance del modelo
performance = model.model_performance(test_data=test)

print(performance)
```
